Q1. Fix the column datatypes

a) Change 'ordered_at' datatype from object to datetime

b) Change 'price' and 'line_total' datatype from object to float

```python
df["ordered_at"] = pd.to_datetime(df["ordered_at"])
for column in ['price', 'line_total']:
    df[column] = df[column].apply(lambda x: pd.to_numeric(x[1:]))
```

Q4. Use regular expression and lambda function to parse data

**TODO**: Get value between "" in name and put it in category column by using .apply

E.g: "ICE CREAM" Peanut Fudge, here we want to extract the phrase "ICE CREAM", and put it in newly created column called category

```python
# Regular expression pattern to extract category and name
pattern = r'^"([A-Z ]+)" (.*)'
transform_func = lambda x: re.findall(pattern, x)[0]

df[['category', 'name']] = df['name'] \
    .apply(transform_func) \
    .apply(pd.Series)
```

Analysis, finally!

```python
# Define colors for each bar
colors = ['skyblue', 'salmon', 'lightgreen', 'purple', 'orange', 'cyan', 'pink', 'yellow', 'green', 'red']
# Plotting
f, ax = plt.subplots(figsize=(10, 6))
df.groupby('name')['line_total'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color=colors)
f.autofmt_xdate()
plt.show()
```
